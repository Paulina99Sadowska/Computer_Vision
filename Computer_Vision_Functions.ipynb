## ReLU Function

$ f = max(x,*0*)$

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set
```

```python
def relu_function(x):
  return max(x,0)
```

```python
for i in [-10,-5,0,5,10]:
  print(relu_function(i))
```

```python
data = np.arange(-5,5,0.1)
mAx = [relu_function(x) for x in data]
plt.plot(mAx)
plt.title('ReLu Function')
```

## ReLU Function

### $ f(x) = \frac{1}{1+e^{-x}}$



```python
def sigmoid_fc(x):
  return 1.0 / (1.0+np.exp(-x))
```

```python
data2 = np.arange(-5, 5, 0.1)
sig = [sigmoid_fc(x) for x in data2]
plt.plot(sig)
plt.title("Sigmoid Function")
```

## Tan h

```python
def tanh(x):
  return (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
```

```python
data = np.arange(-5,5,0.1)
tanh_data = [tanh(x) for x in data]
plt.plot(data,tanh_data)
```
